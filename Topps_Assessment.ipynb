```python
my_list = []

n = int(input("Enter number of elements: "))

for i in range(n):
    a = input("Enter element: ")
    my_list.append(a)

k = int(input("Enter value of k: "))

# Count frequency
freq = {}

for item in my_list:
    if item in freq:
        freq[item] += 1
    else:
        freq[item] = 1

# Print elements with frequency > k
print("Elements with frequency greater than", k, ":")

for key in freq:
    if freq[key] > k:
        print(key)
```

```python
import itertools
import random
comb = []
n = int(input("Enter the number of elements: "))

for i in range(n):
        comb1 = input("Enter the elements: ")
        comb.append(comb1)
print(comb)
population = list(itertools.product(comb))
res = list(itertools.combinations(comb, 3))

life = random.choice(comb)
lifes = random.choices(comb, k = 3)
art = random.sample(comb, k = 3)

print(population)
print(life, lifes)
print(art)
print(res)
```

```python
my_list = []

n = int(input("Enter number of elements: "))

for i in range(n):
    a = int(input("Enter element: "))
    my_list.append(a)

# Square each element
squared_list = []

for item in my_list:
    squared_list.append(item * item)

# Reverse the list
squared_list.reverse()

print("Result:", squared_list)
```

```python
main_list = []

while True:
    choice = input("Enter the type: ")
    
    if choice == "value":
        val = input("Enter the value: ")
        main_list.append(val)
    elif choice == "list":
        temp = []
        n = int(input("Enter the number of element in this list: "))
        
        for i in range(n):
            x = input("Enter the element: ")
            temp.append(x)
        main_list.append(temp)
    elif choice == "empty":
        main_list.append([])
    elif choice == "stop":
        break
    else:
        print("Invalid Input")

print("Final List: ", main_list)

emp = []
for item in main_list:
    if item != []:
        emp.append(item)

print(f"After removing empty lists:", emp)
```
