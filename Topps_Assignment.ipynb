```python
marks = []
n = int(input("Enter the number of dictionaries: "))
if (n<=2):
    print(f"The question cannot process.")
else:
    for i in range(n):
        student = {}
        m = int(input("Enter the number of items: "))
        for j in range(m):
            key = input("Enter the key: ")
            value = input("Enter the values: ")
            student[key] = value

        print("Dictionary", i, "is: ", student)
        marks.append(student)
        print(marks)
        for x in marks:
            print(x)

    comb_student = {}

    for student in marks:

        for key, value in student.items():
            if key in comb_student:
                if type(comb_student[key]) == int:
                    comb_student[key] += int(value)
                else:
                    comb_student[key] += value
            else:
                try:
                    comb_student[key] = int(value)
                except:
                    comb_student[key] = value
    print("The combined dictionary is: ", comb_student)
```

```python
def factorial(n):
    if (n > 0):
        fact = 1
        for i in range(1, n+1):
            fact = fact * i
        return fact
    else:
        print("The number is negative.")

num = int(input("Enter the number: "))

print("Factorial is: ", factorial(num))
```
